## Dataset de Películas

Este dataset contiene información sobre diversas películas, generada de manera aleatoria. Cada fila representa una película con los siguientes atributos:

### Variables:
- **Título**: Nombre de la película.
- **Año**: Año de lanzamiento (entre 2000 y 2025).
- **Género**: Categoría principal de la película (Acción, Comedia, Drama, etc.).
- **Director**: Director de la película.
- **Duración (min)**: Duración de la película en minutos.
- **País**: País de origen de la película.
- **Idioma**: Idioma principal de la película.
- **Rating**: Calificación promedio de la película (entre 8.0 y 10.0).
- **Ingresos Mundiales (millones)**: Ingresos globales en millones de dólares.

Este conjunto de datos es útil para realizar análisis sobre las características y desempeño de las películas en la industria cinematográfica.


```python
import random
import pandas as pd


titulos = [
    "The Shawshank Redemption", "The Godfather", "The Dark Knight", "Pulp Fiction", "Schindler's List",
    "The Lord of the Rings: The Return of the King", "Forrest Gump", "Inception", "Fight Club", "The Matrix",
    "Goodfellas", "The Empire Strikes Back", "Interstellar", "City of God", "Spirited Away",
    "The Green Mile", "Parasite", "The Pianist", "The Departed", "Gladiator",
    "The Lion King", "The Prestige", "Whiplash", "The Intouchables", "The Usual Suspects",
    "Se7en", "Saving Private Ryan", "The Silence of the Lambs", "The Avengers", "Joker",
    "Avengers: Endgame", "Titanic", "The Lord of the Rings: The Fellowship of the Ring", "The Lord of the Rings: The Two Towers", "Star Wars: A New Hope",
    "Back to the Future", "The Social Network", "The Wolf of Wall Street", "Django Unchained", "La La Land",
    "The Truman Show", "The Sixth Sense", "The Grand Budapest Hotel", "Black Panther", "Bohemian Rhapsody",
    "The Imitation Game", "The King's Speech", "Slumdog Millionaire", "A Beautiful Mind", "The Departed"
]


generos = ["Acción", "Comedia", "Drama", "Ciencia ficción", "Terror", "Romance"]
directores = ["Steven Spielberg", "Christopher Nolan", "Martin Scorsese", "Quentin Tarantino", "James Cameron", "Peter Jackson", "Ridley Scott", "Francis Ford Coppola", "David Fincher", "Alfonso Cuarón"]
paises = ["EE.UU.", "Reino Unido", "Francia", "Alemania", "Canadá", "España", "Italia", "Japón", "India", "Australia"]
idiomas = ["Inglés", "Español", "Francés", "Alemán", "Japonés"]


data = []
for titulo in titulos:
    pelicula = {
        "titulo": titulo,
        "año": random.randint(2000, 2025),
        "genero": random.choice(generos),
        "director": random.choice(directores),
        "duracion_min": random.randint(100, 150),
        "pais": random.choice(paises),
        "idioma": random.choice(idiomas),
        "rating": round(random.uniform(8.0, 10.0), 1),
        "ingresos_mundiales_millones": round(random.uniform(1, 6), 2)
    }
    data.append(pelicula)


df = pd.DataFrame(data)




```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear el histograma
plt.figure(figsize=(10, 6))
sns.countplot(x="director", hue="genero", data=df, palette="Set2")

# Títulos y etiquetas
plt.title("Distribución de Géneros por Director", fontsize=16)
plt.xlabel("Director", fontsize=12)
plt.ylabel("Cantidad de Películas", fontsize=12)
plt.xticks(rotation=45)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear el histograma
plt.figure(figsize=(10, 6))
sns.countplot(x="pais", hue="genero", data=df, palette="Set2")

# Títulos y etiquetas
plt.title("Distribución de Géneros por Pais, fontsize=16")
plt.xlabel("Paises", fontsize=12)
plt.ylabel("Cantidad de Películas", fontsize=12)
plt.xticks(rotation=45)

          
# Mostrar gráfico
plt.tight_layout()
plt.show()
```

```python
for director in directores:
    plt.figure(figsize=(10, 6))
    # Filtrar el DataFrame para el director actual
    df_director = df[df["director"] == director]
    
    # Crear el gráfico para ese director
    sns.countplot(x="genero", data=df_director, palette="Set2")
    
    # Títulos y etiquetas
    plt.title(f"Distribución de Géneros para {director}", fontsize=16)
    plt.xlabel("Género", fontsize=12)
    plt.ylabel("Cantidad de Películas", fontsize=12)
    
    # Mostrar gráfico
    plt.tight_layout()
    plt.show()
```

```python
# Boxplot para detectar outliers en ingresos mundiales
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["ingresos_mundiales_millones"], color="skyblue")

plt.title("Boxplot de Ingresos Mundiales (en millones USD)", fontsize=14)
plt.xlabel("Ingresos Mundiales (millones USD)")
plt.grid(True, axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


paises_unicos = df["pais"].unique()

# Crear gráfico de pastel para cada país
for pais in paises_unicos:
    # Filtrar películas del país actual
    df_pais = df[df["pais"] == pais]
    
    # Contar películas por género (o podrías usar director si quieres)
    conteo = df_pais["genero"].value_counts()

    # Crear gráfico de pastel
    plt.figure(figsize=(6, 6))
    plt.pie(conteo, labels=conteo.index, autopct="%1.1f%%", startangle=140, colors=plt.cm.Set3.colors)
    plt.title(f"Distribución de Géneros en {pais}", fontsize=14)
    plt.axis("equal")  # Para que el pastel sea un círculo
    plt.tight_layout()
    plt.show()
```

## Correlación entre Duración, Rating e Ingresos

Este gráfico muestra la **matriz de correlación** entre las variables **duración**, **rating** e **ingresos mundiales** de las películas.

### Características:
- **Colores**: Los colores indican la fuerza y dirección de la correlación: azul para correlaciones positivas.
- **Valores**: Los valores numéricos muestran el **coeficiente de correlación** entre las variables.
- **Formato**: La matriz está anotada con los valores de correlación.

Este gráfico ayuda a visualizar las relaciones entre la duración de las películas, su calificación promedio y los ingresos globales.

```python
# Correlación entre duración, rating e ingresos
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure()
sns.heatmap(df[['duracion_min', 'rating', 'ingresos_mundiales_millones']].corr(), annot=True, cmap="Blues")
plt.title("Matriz de correlación")
plt.xticks(rotation=45)
plt.yticks(rotation=360)
plt.show()
```
